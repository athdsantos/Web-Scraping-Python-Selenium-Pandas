```
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
import pandas as pd
import numpy as np
import requests
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver =webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```
url = 'https://www.aosfatos.org/noticias/checamos/falso/'
```

```
try:
  req = requests.get(url)
  if req.status_code == 200:
    driver.get(url)
    print(f'Status: {req.status_code}, site OK')
  else:
    print(f'Status: {req.status_code}, bad!')
except Exception as error:
  print(error)
```

# Getting infos from news: url, journalist, verdict and url of image

```
driver.get('https://www.aosfatos.org/noticias/checamos/falso/')
```

```
card = driver.find_elements_by_xpath('//*[@id="three-cards"]')
```

```
list_elements_url_etc = []
i = 1
while i <= 15:
  sleep(2)
  link = driver.find_element_by_xpath(f'//*[@id="three-cards"]/a[{i}]').get_attribute('href')
  
  sleep(2)
  driver.execute_script("window.scrollBy(0,350)")
  sleep(2)
  
  driver.find_element_by_xpath(f'//*[@id="three-cards"]/a[{i}]').click()
  sleep(2)

  article = driver.find_element_by_tag_name('article')
  sleep(3)
  
  author = article.find_element_by_class_name('author')
  sleep(3)
  
  try:
    verdict = article.find_element_by_name('e5e7')
  except:
    all_p = driver.find_elements_by_tag_name('p')
    for p in all_p:
      if p.text == 'FALSO'.upper():
        verdict = p
  
  p_img = article.find_elements_by_tag_name('img')
  try:
    p_img = article.find_elements_by_tag_name('img')
    list_elements_url_etc.append([link, author.text[4:], verdict.text, p_img[1].get_attribute('src')])
  except IndexError:
    for p in p_img:
      if 'medium' in p.get_attribute('src'):
        list_elements_url_etc.append([link, author.text[4:], verdict.text, 'EMPTY'])
  except Exception as erro:
    print(erro)
  
  sleep(2)
  driver.back()
  sleep(2)
  #print(i, link)
  i += 1
```

```
for element in list_elements_url_etc:
  print(f'Link: {element[0]}\nJournalist: {element[1]}\nVerdict: {element[2]}\nURL Image: {element[3]}\n\n')
```

# Getting title, date, link, author, verdict and url from image from the next three pages.

```
#coloque seu codigo aqui
driver.get('https://www.aosfatos.org/noticias/checamos/falso/')
```

```
card = driver.find_elements_by_xpath('//*[@id="three-cards"]')
list_final = []
```

```
cont_page = 1
while cont_page <= 3:
  i = 1
  while i <= 15:
    sleep(2)
    link = driver.find_element_by_xpath(f'//*[@id="three-cards"]/a[{i}]').get_attribute('href')
    
    sleep(2)
    #list_final.append([link])
    driver.execute_script("window.scrollBy(0,750)")
    sleep(2)
    
    driver.find_element_by_xpath(f'//*[@id="three-cards"]/a[{i}]').click()
    sleep(3)

    article = driver.find_element_by_tag_name('article')
    sleep(2)

    title_news = article.find_element_by_tag_name('h1')
    sleep(3)

    data = article.find_element_by_class_name('publish_date')
    sleep(3)

    author = article.find_element_by_class_name('author')
    sleep(3)
    
    try:
      verdict = article.find_element_by_name('e5e7')
    except:
      all_p = driver.find_elements_by_tag_name('p')
      for p in all_p:
        if p.text == 'FALSO'.upper():
          verdict = p
    
    p_img = article.find_elements_by_tag_name('img')
    try:
      p_img = article.find_elements_by_tag_name('img')
      url_img = p_img[1].get_attribute('src')
      list_final.append([title_news.text, data.text, link, author.text[4:], verdict.text, url_img])
    except IndexError:
      for p in p_img:
        if 'medium' in p.get_attribute('src'):
          url_img = 'EMPTY'
          list_final.append([title_news.text, data.text, link, author.text[4:], verdict.text, url_img])
    except Exception as erro:
      print(erro)
    
    sleep(2)
    driver.back()
    sleep(2)
    print(cont_page, i, link)
    i += 1
  driver.execute_script("window.scrollBy(0,900)")
  sleep(2)
  if cont_page == 3:
    break
  else:
    buttonNext = driver.find_element_by_xpath(f'/html/body/section[2]/div/a[{cont_page}]').click()
    sleep(2)
    cont_page += 1
    sleep(2)
```

```
for l in list_final:
  print(l)
```

# Saving informations in a DataFrame using Pandas lib.

```
#coloque seu código aqui
df_news = pd.DataFrame(data=list_final, columns=['Title', 'Date', 'Link', 'Author', 'Verdict', 'URL Image'])
```
